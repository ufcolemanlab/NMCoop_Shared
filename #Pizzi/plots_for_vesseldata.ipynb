```python
import numpy as np
import matplotlib.pyplot as plt

def plot_bar_with_sem(data1, data2, font_style='Arial', font_size=12):
    # Calculate mean and SEM for each group
    mean1 = np.mean(data1)
    sem1 = np.std(data1) / np.sqrt(len(data1))
    mean2 = np.mean(data2)
    sem2 = np.std(data2) / np.sqrt(len(data2))

    # Plot means with error bars (SEM)
    bar_width = 0.35
    index = np.arange(2)
    bars = plt.bar(index, [mean1, mean2], bar_width, yerr=[sem1, sem2], color=['blue', 'green'], alpha=0.7)

    # Set font style and size for all labels
    plt.rcParams['font.family'] = font_style
    plt.rcParams['font.size'] = font_size

    plt.xlabel('Groups')
    plt.ylabel('Values')
    plt.xticks(index, ['Group 1', 'Group 2'])

    # Add legends for both groups
    plt.legend(bars, ['Group 1', 'Group 2'], frameon=False, loc='best')
    
    plt.title('Bar Plot of Means Â± SEM for Two Groups')
    plt.show()


```

```python
#As a function:

import numpy as np
import matplotlib.pyplot as plt

def plot_vertical_histograms(group1_data, group2_data, bins=30, figsize=(8, 8), font_type='Arial', font_size=12):
    """
    Plot histograms of two groups vertically with Group 1 histogram flipped along the vertical axis.

    Parameters:
    - group1_data: Array-like, data for Group 1.
    - group2_data: Array-like, data for Group 2.
    - bins: int or sequence of scalars or str, optional (default=30)
            If an integer, defines the number of equal-width bins in the range.
            If a sequence, defines the bin edges, including the rightmost edge, allowing for non-uniform bin widths.
            If a string, it defines the method used to calculate the optimal bin width, as defined by `np.histogram_bin_edges`.
    - figsize: tuple, optional (default=(8, 8))
               Width, height in inches.
    - font_style: str, optional (default='normal')
                  Font style.
    - font_size: int or float, optional (default=12)
                 Font size.

    Returns:
    - fig: Figure object
    - axs: Array of Axes objects
    """
    # Set font style and size
    plt.rcParams.update({'font.family': font_type, 'font.style': 'normal', 'font.size': font_size})

    # Calculate the bin edges
    hist_range = (min(min(group1_data), min(group2_data)), max(max(group1_data), max(group2_data)))
    bins = np.linspace(*hist_range, bins)

    # Create subplots
    fig, axs = plt.subplots(2, 1, figsize=figsize, sharex=True)

    # Plot histogram for Group 2 on the first subplot
    axs[0].hist(group2_data, bins=bins, color='green', alpha=0.5, orientation='horizontal')
    axs[0].set_ylabel('Values')
    axs[0].set_title('Group 2')

    # Plot histogram for Group 1 on the second subplot, flipped along the vertical axis
    axs[1].hist(group1_data, bins=bins, color='blue', alpha=0.5, orientation='horizontal')
    axs[1].invert_xaxis()
    axs[1].set_xlabel('Frequency')
    axs[1].set_ylabel('Values')
    axs[1].set_title('Group 1')

    plt.tight_layout()
    plt.show()

    return fig, axs

# Example usage:
group1_data = np.random.normal(loc=0, scale=1, size=1000)
group2_data = np.random.normal(loc=2, scale=1.5, size=1000)
plot_vertical_histograms(group1_data, group2_data, figsize=(10, 8), font_type='Arial', font_size=14)


"""This function plot_vertical_histograms takes in the data for Group 1 and Group 2, and
optionally the number of bins, and plots the histograms vertically with Group 1 flipped
along the vertical axis. It returns the fig and axs objects for further customization if needed."""
```

```python
import numpy as np
import matplotlib.pyplot as plt

def plot_cumulative_percent(group1_data, group2_data, bins=30, figsize=(8, 6), font_style='normal', font_size=12):
    """
    Plot cumulative percent histograms of two groups of input data.

    Parameters:
    - group1_data: Array-like, data for Group 1.
    - group2_data: Array-like, data for Group 2.
    - bins: int or sequence of scalars or str, optional (default=30)
            If an integer, defines the number of equal-width bins in the range.
            If a sequence, defines the bin edges, including the rightmost edge, allowing for non-uniform bin widths.
            If a string, it defines the method used to calculate the optimal bin width, as defined by `np.histogram_bin_edges`.
    - figsize: tuple, optional (default=(8, 6))
               Width, height in inches.
    - font_style: str, optional (default='normal')
                  Font style.
    - font_size: int or float, optional (default=12)
                 Font size.
    """
    # Set font style and size
    plt.rcParams.update({'font.family': 'Arial', 'font.style': font_style, 'font.size': font_size})

    # Calculate the bin edges
    hist_range = (min(min(group1_data), min(group2_data)), max(max(group1_data), max(group2_data)))
    bins = np.linspace(*hist_range, bins)

    # Create subplots
    fig, ax = plt.subplots(figsize=figsize)

    # Plot cumulative percent histograms for both groups
    weights1 = np.ones_like(group1_data) / len(group1_data)
    weights2 = np.ones_like(group2_data) / len(group2_data)

    ax.hist(group1_data, bins=bins, color='blue', alpha=0.5, cumulative=True, weights=weights1, label='Group 1', histtype='step', linewidth=2)
    ax.hist(group2_data, bins=bins, color='green', alpha=0.5, cumulative=True, weights=weights2, label='Group 2', histtype='step', linewidth=2)

    # Set labels and title
    ax.set_xlabel('Values')
    ax.set_ylabel('Cumulative Percent')
    ax.set_title('Cumulative Percent Histograms of Group 1 and Group 2')

    # Add legend
    ax.legend()

    plt.show()

# Example usage:
group1_data = np.random.normal(loc=0, scale=1, size=1000)
group2_data = np.random.normal(loc=2, scale=1.5, size=1000)
plot_cumulative_percent(group1_data, group2_data, figsize=(10, 8), font_style='italic', font_size=14)
```
