```python
# function new - new df for box plots etc

def filter_dataframe(df, metric_prefix, second_metric_prefix, logical_prefix, group, vessel_type):
    """
    Filters the DataFrame to return all relevant rows for a specific group and metric columns 
    based on corresponding logical columns being 1, and assigns vessel_type.

    Parameters:
    df (pd.DataFrame): The input DataFrame.
    metric_prefix (str): The prefix for the first metric column (e.g., 'meanFWHM').
    second_metric_prefix (str): The prefix for the second metric column (e.g., 'meanFeFv').
    logical_prefix (str): The prefix for logical columns (e.g., 'capLogical').
    group (str): The group ID (e.g., 'A', 'B') for row filtering.
    vessel_type (str): The vessel type to assign (e.g., 'capillary', 'venule').

    Returns:
    pd.DataFrame: A filtered DataFrame with all metric columns, group, vessel_type.
    """
    # Filter rows based on the group
    filtered_df = df[df['group'] == group]

    # Get metric and logical columns
    metric_columns = [col for col in df.columns if col.startswith(metric_prefix)]
    second_metric_columns = [col for col in df.columns if col.startswith(second_metric_prefix)]
    logical_columns = [col for col in df.columns if col.startswith(logical_prefix)]

    # Initialize result DataFrame to store all relevant rows
    results = []

    # For each logical column, check where the logical is 1 and gather metrics for those rows
    for logical_col in logical_columns:
        condition_df = filtered_df[filtered_df[logical_col] == 1]

        # Gather corresponding metric columns
        if not condition_df.empty:
            for metric_col, second_metric_col in zip(metric_columns, second_metric_columns):
                result_part = pd.DataFrame({
                    'meanFWHM': condition_df[metric_col],
                    'meanFeFv': condition_df[second_metric_col],
                    'group': group,
                    'vessel_type': vessel_type,
                    'filename': condition_df['filename']
                })
                results.append(result_part)

    # Concatenate all the parts and return as a single DataFrame
    return pd.concat(results, ignore_index=True) if results else pd.DataFrame()

# EXAMPLE USE
# Filter for capillary in group A
filtered_resultA_cap = filter_dataframe(df, 'meanFWHM', 'meanFeFv', 'capLogical', 'A', 'capillary')

# Filter for venule in group A
filtered_resultA_ven = filter_dataframe(df, 'meanFWHM', 'meanFeFv', 'venLogical', 'A', 'venule')

# Filter for arteriole in group A
filtered_resultA_art = filter_dataframe(df, 'meanFWHM', 'meanFeFv', 'artLogical', 'A', 'arteriole')

# Similarly for group B
filtered_resultB_cap = filter_dataframe(df, 'meanFWHM', 'meanFeFv', 'capLogical', 'B', 'capillary')
filtered_resultB_ven = filter_dataframe(df, 'meanFWHM', 'meanFeFv', 'venLogical', 'B', 'venule')
filtered_resultB_art = filter_dataframe(df, 'meanFWHM', 'meanFeFv', 'artLogical', 'B', 'arteriole')

# Concatenate all filtered results into one dataframe
final_df = pd.concat([filtered_resultA_cap, filtered_resultB_cap, 
                      filtered_resultA_ven, filtered_resultB_ven, 
                      filtered_resultA_art, filtered_resultB_art], ignore_index=True)

print(final_df.head())


# # Filter for vessel type in group A
# filtered_resultA_cap = filter_dataframe(df, 'meanFWHM', 'meanFeFv', 'capLogical', 'A', 'capillary')
# filtered_resultA_ven = filter_dataframe(df, 'meanFWHM', 'meanFeFv', 'venLogical', 'A', 'venule')
# filtered_resultA_art = filter_dataframe(df, 'meanFWHM', 'meanFeFv', 'artLogical', 'A', 'arteriole')

# # Filter for vessel type in group B
# # filtered_resultB_cap = filter_dataframe(df, 'meanFWHM', 'meanFeFv', 'capLogical', 'B', 'capillary')
# filtered_resultB_cap = filter_dataframe(df, 'meanFWHM', 'meanFeFv', 'capLogical', 'B', 'capillary')
# filtered_resultB_ven = filter_dataframe(df, 'meanFWHM', 'meanFeFv', 'venLogical', 'B', 'venule')
# filtered_resultB_art = filter_dataframe(df, 'meanFWHM', 'meanFeFv', 'artLogical', 'B', 'arteriole')

# # print(filtered_resultA_cap.head())
# # print(filtered_resultA_ven.head())
# # print(filtered_resultA_art.head())

print(final_df['meanFWHM'])
print(final_df['group'])
print(final_df['vessel_type'])


# You can continue for other vessel types (e.g., venuleLogical, artLogical)
```

```python
import pandas as pd
from sklearn.decomposition import PCA
from sklearn.cluster import KMeans
# import umap
from sklearn.impute import SimpleImputer

# Sample dataframe structure for illustration
# df = pd.read_csv('analysisData.csv')

# Step 1: Dynamically identify columns that contain the relevant substrings
mean_cols = [col for col in df.columns if 'meanFWHM' in col]
stdev_cols = [col for col in df.columns if 'stdevFWHM' in col]
min_cols = [col for col in df.columns if 'minFWHM' in col]
max_cols = [col for col in df.columns if 'maxFWHM' in col]

# Combine all relevant columns
relevant_columns = mean_cols + stdev_cols + min_cols + max_cols

# Step 2: Track rows before filtering
df_before = df.copy()

# Step 3: Drop rows with any of these conditions:
# 1. 'roiName' is 'Background'
# 2. Row contains NaN in any of the relevant columns
# df_clean = df[~((df['roiName'] == 'Background') | df[relevant_columns].isna().any(axis=1))]
df_clean = df[~((df['roiName'] == 'Background') & 
               df['roiName'].duplicated(keep=False))]

# Track what was dropped
dropped_rows = df_before[~df_before.index.isin(df_clean.index)]
print("Dropped rows:")
print(dropped_rows[['roiID', 'filename', 'roiName']])

# Step 4: For remaining rows, replace NaNs in relevant columns with the row's overall mean
for col in relevant_columns:
    df_clean[col] = df_clean[col].fillna(df_clean[relevant_columns].mean(axis=1))

# print("Cleaned up data:")
# print(df_clean)

# Step 5: Perform PCA (after handling NaNs)
pca = PCA(n_components=2)
pca_result = pca.fit_transform(df_clean[relevant_columns])

# Step 6: KMeans clustering
kmeans = KMeans(n_clusters=2)
kmeans_result = kmeans.fit_predict(pca_result)

# Step 7: UMAP
# umap_model = umap.UMAP(n_neighbors=5)
# umap_result = umap_model.fit_transform(df_clean[relevant_columns])

# Step 8: Add results back to the clean dataframe
df_clean['PCA1'] = pca_result[:, 0]
df_clean['PCA2'] = pca_result[:, 1]
df_clean['KMeans_cluster'] = kmeans_result
# df_clean['UMAP1'] = umap_result[:, 0]
# df_clean['UMAP2'] = umap_result[:, 1]

# Step 9: View results including 'roiID' and 'filename'
#print(df_clean[['roiID', 'filename', 'PCA1', 'PCA2', 'KMeans_cluster', 'UMAP1', 'UMAP2']].head())
print(df_clean[['roiID', 'roiName''filename', 'PCA1', 'PCA2', 'KMeans_cluster']].head())
```
